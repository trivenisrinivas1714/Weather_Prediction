```python
#NaiveBayes project (Weather Prediction)
#Required Modules
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
#Reading CSV files
df = pd.read_csv("seattle-weather.csv")
df
```

```python
#Encoding the strings to Numericals
outlook_at=LabelEncoder()
Temp_at=LabelEncoder()
Hum_at=LabelEncoder()
win_at=LabelEncoder()
```

```python
#Dropping the target variable and make it is as newframe
inputs=df.drop('weather',axis='columns')
target=df['weather']
target
```

```python
#Creating the new dataframe
inputs['date_n']= outlook_at.fit_transform(inputs['date'])
inputs['precipitation_n']= outlook_at.fit_transform(inputs['precipitation'])
inputs['temp_max_n']= outlook_at.fit_transform(inputs['temp_max'])
inputs['temp_min_n']= outlook_at.fit_transform(inputs['temp_min'])
inputs['wind_n']= outlook_at.fit_transform(inputs['wind'])
inputs
```

```python
#Dropping the string values
inputs_n=inputs.drop(['date','precipitation','temp_max','temp_min'],axis='columns')
inputs_n
```

```python
#Applying the Gaussian naivebayes
classifier = GaussianNB()
classifier.fit(inputs_n,target)
```

```python
#85% accuracy 
classifier.score(inputs_n,target)
```
